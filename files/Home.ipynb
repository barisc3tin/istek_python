# İSTEK Python Notebook Portalı

Bu portal, 10. sınıf Python kodlama görevleri için hazırlanmıştır.

## Kullanım
1) İlgili haftaya tıklayın ve notebook’u açın.  
2) File → **Save As…** ile dosyanızı şu formatta kaydedin:  
   **10A_AdSoyad_21Hafta.ipynb**  
3) Görevleri tamamlayın.  
4) File → **Download → Notebook (.ipynb)**  
5) File -> **Save as -> PDF**
5) Öğretmeninize gönderin.
