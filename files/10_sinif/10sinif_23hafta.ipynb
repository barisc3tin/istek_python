## 4. `print` ve `return` farkı

- `print()` sadece ekranda gösterir.
- `return` ise sonucu geri döndürür ve programın başka bir yerinde kullanmamızı sağlar.

````python
def topla_yaz(a, b):
    print(a + b)

def topla_dondur(a, b):
    return a + b

topla_yaz(3, 5)

sonuc = topla_dondur(3, 5)
print("Dönen değer:", sonuc)
print("Sonucun 2 katı:", sonuc * 2)
````
